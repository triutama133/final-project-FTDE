```python
#!python3

import json
import time
import kafka
import six
import pandas as pd

from kafka import KafkaProducer

def json_serializer(data):
    return json.dumps(data).encode("utf-8")

if __name__ == "__main__":

    data = pd.read_csv('/content/drive/MyDrive/Final Project/data_recruitment_selection.csv')
    json_data = data.to_dict(orient='records')

    producer = KafkaProducer(bootstrap_servers=['34.142.138.97'], value_serializer=json_serializer)

    # Loop through the data once, then break
    for data in json_data:
        print(data)
        producer.send("ftde02-finalproject-kel5-fix", data)
```
